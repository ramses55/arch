```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
images = ['../Data/528.jpeg', '../Data/1084.jpg', '../Data/12146_И6125.jpg', '../Data/1753.jpeg', '../Data/12245_И6774.jpg',
          '../Data/12145_И6666.jpg', '../Data/1038.jpg']
image = images[1]
```

```python
img = cv2.imread(image)

plt.imshow(cv2.blur(cv2.cvtColor(img, cv2.COLOR_BGR2RGB), (21,21)))
plt.show()
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)


lower_bound = (0,0,168)
upper_bound = (172,111,255)

mask = cv2.inRange(img_hsv, lower_bound, upper_bound)

plt.imshow(mask)
plt.show()
```
